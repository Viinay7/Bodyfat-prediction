```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("C:/Users/viina/Downloads/bodyfat.csv")
data = {
    "Age": [25, 30, 22, 35, 28, 45, 50, 33, 40, 29, 32, 31, 37, 43, 48, 39, 41, 36],
    "BodyFat": [20.7, 25.0, 18.9, 26.4, 23.5, 30.2, 28.0, 24.1, 27.6, 22.3, 24.9, 23.0, 26.8, 29.5, 27.7, 25.4, 28.9, 26.1]
}

df = pd.DataFrame(data)

# a) Find the maximum and the minimum percentage of the fat and age of the adults who visited the hospital
max_age = df['Age'].max()
min_age = df['Age'].min()

max_bodyfat = df['BodyFat'].max()
min_bodyfat = df['BodyFat'].min()

# b) Calculate mean, median and midrange of the age
mean_age = df['Age'].mean()
median_age = df['Age'].median()
midrange_age = (df['Age'].max() + df['Age'].min()) / 2

# c) Find the first quartile and third quartile of the data
first_quartile_age = df['Age'].quantile(0.25)
third_quartile_age = df['Age'].quantile(0.75)

# d) Draw a scatter plot and q-q plot based on these two variables
plt.figure(figsize=(14, 7))

# Scatter plot
plt.subplot(1, 2, 1)
sns.scatterplot(x='Age', y='BodyFat', data=df)
plt.title('Scatter Plot of Age vs Body Fat')
plt.xlabel('Age')
plt.ylabel('Body Fat')

# Q-Q plot
plt.subplot(1, 2, 2)
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Age')

plt.tight_layout()
plt.show()

# Output the results
print(f"Maximum age: {max_age}")
print(f"Minimum age: {min_age}")
print(f"Maximum body fat percentage: {max_bodyfat}")
print(f"Minimum body fat percentage: {min_bodyfat}")
print(f"Mean age: {mean_age}")
print(f"Median age: {median_age}")
print(f"Midrange age: {midrange_age}")
print(f"First quartile age: {first_quartile_age}")
print(f"Third quartile age: {third_quartile_age}")
```
